```python
df.columns=['State','Total_cases','Active','Death','Active_Ratio','Death_Ratio']
df.head(3)
```

```python
df.groupby(['State'])['Active_Ratio'].agg(['max','min'])
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['State'],df['Active_Ratio'],color='red',marker='o')
plt.xticks(rotation=90,fontsize=13.5)
plt.yticks(fontsize=4)
plt.xlabel('State Names')
plt.ylabel('Active_Ratio')
plt.title('Line plot for active ratios cases')
plt.show()
```

```python
Total_cases_state=df.groupby('State')['Total_cases'].sum().sort_values(ascending=False)
Total_cases_state
```

```python
plt.figure(figsize=(10,4))
plt.bar(df['State'],df['Total_cases'])
plt.xticks(rotation=90)
plt.title('Total Cases in Each State',fontsize=20)
plt.xlabel('States Name')
plt.ylabel('Total Case')
plt.show()
```
