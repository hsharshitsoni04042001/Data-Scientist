# Renaming the column's names for better readability

```python
df.columns=["state","total_cases","active_cases","deaths","active_ratio","death_ratio"]
df.head(5)
```

# Aggregating state wise active ratios

```python
df.groupby(['state'])['active_ratio'].agg(['max','min'])
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['state'],df['active_ratio'],color='red',marker='o')
plt.xticks(rotation=90,fontsize=13.5)
plt.yticks(fontsize=13.5)
plt.xlabel('state',fontsize=15)
plt.ylabel('active_ratio',fontsize=15)
plt.title('Line Plot for Active_ratio',fontsize=15)
plt.show()
```

# Aggregating state wise death ratios

```python
df.groupby(['state'])['death_ratio'].agg(['max','min'])
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['state'],df['death_ratio'],color="black",marker='o')
plt.xticks(rotation=90,fontsize=13.5)
plt.yticks(fontsize=13.5)
plt.xlabel('state',fontsize=10)
plt.ylabel('death_ratio',fontsize=10)
plt.title('"Line Plot for State wise Death_ratio"')
plt.show()
```

# Calculating total cases by aggregating state wise total cases

```python
total_cases_state=df.groupby(['state'])['total_cases'].sum().sort_values(ascending=False)
total_cases_state
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
sns.barplot(total_cases_state.index,total_cases_state)
plt.title('"Total cases in each state"',fontsize=20)
plt.xlabel('states',fontsize=15)
plt.ylabel('total_cases',fontsize=15)
plt.show()
```

# Top 10 states with Total no of cases

```python
top_10=total_cases_state[:10]
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
sns.barplot(top_10.index,top_10)
plt.title('"Top 10 total cases states"',fontsize=20)
plt.xlabel('States',fontsize=20)
plt.ylabel('Total cases',fontsize=15)
plt.show()
```

# Creating new column 'Recovered_cases' (Feature engineering)

```python
recovered_cases=df['total_cases']-(df['active_cases']+df['deaths'])
df['Recovered_cases']=recovered_cases
df.head(5)
```

# Calculating Recovered cases by aggregating state wise recovered cases

```python
rec_cases=df.groupby(['state'])['Recovered_cases'].sum().sort_values(ascending=False)
rec_cases
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
plt.plot(df['state'],df['Recovered_cases'],marker='o',color='green')
plt.title('"Total Recovered cases in each state"',fontsize=20)
plt.xlabel('states',fontsize=15)
plt.ylabel('Recovered_cases',fontsize=15)
plt.show()
```
